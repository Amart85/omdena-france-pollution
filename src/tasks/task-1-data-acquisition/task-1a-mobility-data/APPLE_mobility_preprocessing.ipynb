## Plotting the Apple Mobility Data for a Specific City
Source: https://www.viralml.com/video-content.html?v=A6nWZKZpL3k

```python
paris_df = apple_fr_df[ apple_fr_df['region'] == 'Paris' ]
paris_df
```

```python
for i,row in paris_df.iterrows():
    fig, ax = plt.subplots(figsize=(16, 4))
    
    title = row[0] + row[1] + row[2]
    df = pd.DataFrame({'Date':row.index[6:], 'Activity':[float(f) for f in list(row.values[6:])]})
    df['Date'] = pd.to_datetime(df['Date'])
        
    plt.title(title)
    plt.grid(alpha=0.5)
    plt.plot(df['Date'], df['Activity'])

    # 17 March 2020 – 10 May 2020: First national lockdown
    ax.axvline(pd.to_datetime('2020-03-17'), color='red', alpha=0.5, linestyle='--', lw=2)
    plt.text(pd.to_datetime('2020-03-23'), 60, '1st National Lockdown', color='r', rotation=90)
    ax.axvline(pd.to_datetime('2020-05-10'), color='red', alpha=0.5, linestyle='--', lw=2)


    # 30 October 2020 – 14 December 2020: Second national lockdown
    ax.axvline(pd.to_datetime('2020-10-30'), color='red', alpha=0.5, linestyle='--', lw=2)
    plt.text(pd.to_datetime('2020-11-05'), 60, '2nd National Lockdown', color='r', rotation=90)
    ax.axvline(pd.to_datetime('2020-12-14'), color='red', alpha=0.5, linestyle='--', lw=2)
```
