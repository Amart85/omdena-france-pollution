# Particulate Matter

## Particulate Matter 2.5µm

Definitions : https://www.epa.gov/pm-pollution/particulate-matter-pm-basics

```python
pm25 = pd.read_csv("2020_PM25.csv")
pm25.head()
```

- AUT : AUTEUIL (Périphérique)
- PA01H : Paris 1er Les Halles
- RN6 : Route Nationale 6 - Melun
- RUR_N : Saint Martin du Tertre
- RUR_SE : Fôret de Fontainebleau
- GON : Gonesse
- A1 : Autoroute A1 - Saint-Denis
- BP_EST : Boulevard Périphérique Est
- RUR_S : Bois Herpin
- BOB : Bobigny
- GEN : Gennevilliers
- VITRY : Vitry-sur-Seine
- RAMBO : Rambouillet

```python
pm25.drop("OBJECTID", axis=1, inplace=True)
```

```python
pm25.columns = ["Date", "AUT", "PA01H", "RN6", "RUR_N", "RUR_SE", "GON", "A1", "BP_EST", "RUR_S", "BOB", "GEN", "VITRY", "RAMBO"]
```

```python
pm25 = pm25.drop([0, 1, 2, 3, 4], axis=0)
pm25.head()
```

```python
pm25.to_csv("datasets_ileDeFrance/pm25_IleDeFrance.csv", index=False)
```
